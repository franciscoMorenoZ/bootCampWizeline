```python
#Q4. Show the 9 Toyota cars with the most extreme Fuel Barrels/Year in abosolute terms within all Toyota cars. 
#Show the car Model, Year and their Fuel Barrels/Year in standard deviation units(Z-score) sorted in descending 
#order by their Fuel Barrels/Year in absolute terms first and then by year in descending order 
#BUT without modifying the negative values (see example).
```

```python
toyota_df = df[df.Make == "Toyota"].copy()
toyota_df.info()
```

```python
toyota_df_sub_set = toyota_df[['Model','Year','Fuel Barrels/Year']].copy()
toyota_df_sub_set
```

```python
#data_z = (toyota_df['Fuel Barrels/Year']-toyota_df['Fuel Barrels/Year'].mean())/(toyota_df['Fuel Barrels/Year'].std())
#data_z = (toyota_df-toyota_df.mean())/(toyota_df.std())
toyota_df_sub_set['Zscore'] = stats.zscore(toyota_df_sub_set['Fuel Barrels/Year'])
toyota_df_sub_set_sorted = toyota_df_sub_set.sort_values(['Zscore','Year'],ascending=True)
answer4 = toyota_df_sub_set_sorted.head(9).set_index('Model')
print(answer4)
```

```python
answer_dict["Q4"] =  answer4
answer_dict
```

```python
#Q5. Calculate the changes in Combined MPG with their previous model of all Golf cars with Manual 
#5-spd transmission and Regular Fuel Type. Show the Year, the Combined MPG and the calculated difference of MPG 
#in a list sorted by Year in ascending order.
```

```python
volkswagen_df2 = (df[(df['Make'] == "Volkswagen")  & (df['Model'] == "Golf") & (df['Transmission'] == "Manual 5-spd") &(df['Fuel Type'] == "Regular")   ].copy())
volkswagen_df2_filtered = volkswagen_df2[['Year','Combined MPG']].copy()
volkswagen_df2_filtered.set_index('Year')
```

```python
volkswagen_df2_filtered['Difference'] = volkswagen_df2_filtered['Combined MPG'].diff()
answer5 = volkswagen_df2_filtered.fillna(0.0).set_index('Year').head(19)
answer5
```

```python
answer_dict["Q5"] =  answer5
answer_dict
```

```python
#Q6. What are the top 5 lowest CO2 Emission Grams/Mile emmisions of cars for each of the following brands: 
#Toyota, Ford, Volkswagen, Nissan, Honda
Brands_df = (df[(df['Make']=='Toyota') | (df['Make']=='Ford') | (df['Make']=='Volkswagen') | (df['Make']=='Nissan') | (df['Make']=='Honda')]).sort_values('CO2 Emission Grams/Mile',ascending=True)[['Make','CO2 Emission Grams/Mile']]
answer6 = (Brands_df.groupby(by='Make').apply(lambda x:x.nsmallest(5,['CO2 Emission Grams/Mile']))).set_index('Make').copy()
answer6
```

```python
answer_dict["Q6"] =  answer6
answer_dict
```
